**利用order()函數,選一種類型的資料出來,由大到小排序**

```
NBA1819Order<-NBA1819[order(NBA1819$FieldGoalsAttempted,decreasing = T),]
NBA1819Order[1:10,]
```

```
NBA1819[grepl("C",NBA1819$Position),]
```

**用ggplot畫出,四種不同類型的圖表**

```
NBA1819<-fetch_NBAPlayerStatistics("18-19")
df<-data.frame(gp=NBA1819$TotalMinutesPlayed,y=NBA1819$TotalPoints)
```

**散佈圖:**

個人犯規與阻攻

```
qplot(PersonalFouls,Blocks,data=NBA1819)
```

上場時長與總得分

```
ggplot(df,aes(x=gp,y=y))+geom_point()
```

**直方圖:**

犯規數

```
qplot(PersonalFouls,data=NBA1819)
```

**折線圖**

```
ggplot(df,aes(NBA1819$TotalMinutesPlayed,NBA1819$TotalPoints,group=1))+geom_line()
```

**盒狀圖**

上場時長與總得分

```
ggplot(df,aes(x=gp,y=y,group=1))+geom_boxplot()
```
